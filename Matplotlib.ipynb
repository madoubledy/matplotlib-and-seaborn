# UNIVARIATE ANALYSIS

```python
df_Europe=df.loc[df['WHO Region']=='Europe']
df_Africa=df.loc[df['WHO Region']=='Africa']
df_EM=df.loc[df['WHO Region']=='Eastern Mediterranean']
```

```python
plt.plot(df_Europe['New cases'],np.ones_like(df_Europe['New cases']),'o')
plt.plot(df_Africa['New cases'],np.ones_like(df_Africa['New cases']),'o')
plt.plot(df_EM['New cases'],np.ones_like(df_EM['New cases']),'o')
plt.xlabel('New Cases')
plt.show()
```

# Bivariate Analysis

```python
sns.FacetGrid(df,hue="WHO Region",size=5).map(plt.scatter,"New cases","New deaths").add_legend();
plt.show()
```
